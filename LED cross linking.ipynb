```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
```

```python
# this notebook is to estimate the power density of our 365nm LED
# We are using the Mouser LZ1-00UV00 that has a flux output of 1200mW at 0.7A
# I measured the diameter of the spot as function of distance
height = np.array([1.8,6.3,11.8])
diameter = np.array([6,12,20])
p = np.polyfit(x=height,y=diameter,deg =1)
print(p)
```

```python
plt.plot(height,diameter,"o")
plt.plot(height,np.poly1d(p)(height))
```

```python
# we want to predict at what height do we get what flux density
led_dens=1200 # in mW
x = np.linspace(0,20)
d = np.poly1d(p)(x)
A = np.pi/4.0*d**2
pd = led_dens/A
plt.plot(x,pd)
plt.xlim((5,20))
plt.ylim((0,20))
plt.xlabel('distance in cm')
plt.ylabel('flux density mW/cm^2')
```

```python
print('distance','diameter','power density')
for xx,dd,pd in zip(x,d,pd):
    print(xx,dd,pd)
```
